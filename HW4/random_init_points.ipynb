```python
import random
from plyfile import PlyData, PlyElement
import numpy as np


def randomize_point_3d():
    # Random coordinates (x, y, z)
    x = random.uniform(-10, 10)
    y = random.uniform(-10, 10)
    z = random.uniform(-10, 10)
    
    # Random normals (nx, ny, nz) and normalize
    nx = random.uniform(-1, 1)
    ny = random.uniform(-1, 1)
    nz = random.uniform(-1, 1)
    
    norm = (nx**2 + ny**2 + nz**2)**0.5
    if norm > 0:
        nx /= norm
        ny /= norm
        nz /= norm
    
    # Random colors (red, green, blue)
    red = random.randint(0, 255)
    green = random.randint(0, 255)
    blue = random.randint(0, 255)
    
    print(x, y, z, nx, ny, nz, red, green, blue)

    return (x, y, z, nx, ny, nz, red, green, blue)

# Generate 13,414 random points
point_3d_data = []
for _ in range(13414):
    point_3d_data.append(randomize_point_3d())

# Convert point_3d data to a structured NumPy array
point_3d_array = np.array(point_3d_data, dtype=[
    ('x', 'f4'), ('y', 'f4'), ('z', 'f4'),  # Position
    ('nx', 'f4'), ('ny', 'f4'), ('nz', 'f4'),  # Normals
    ('red', 'u1'), ('green', 'u1'), ('blue', 'u1')  # Colors
])

# Create PlyElement for the point_3d data (no need to flatten)
point_3d_element = PlyElement.describe(point_3d_array, 'vertex')

# Write the PLY data to file
ply_data = PlyData([point_3d_element])
ply_data.write('randomized_points.ply')

print("PLY file saved as 'randomized_points.ply'")
```

```python
ply_data = PlyData.read('randomized_points.ply')
for element in ply_data.elements:
    print(f"Element name: {element.name}")
    for property in element.properties:
        print(f"Property: {property.name}")

point_3d_data = ply_data['vertex']
print(f"Number of gaussain points: {len(point_3d_data)}")
```
